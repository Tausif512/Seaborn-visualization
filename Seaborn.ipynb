```python
import seaborn as sns
```

```python
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset("fmri") #In seaborn fmri dataset is in-build dataset.
```

```python
fmri.head()
```

**SEABORN LINEPLOT**

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
fmri.shape
```

  **Grouping data with 'hue'**

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

**Adding style**

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event')
plt.show()
```

**Adding markers**

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event', style='event', markers=True)
plt.show()
```

**Histogram/distplot**

```python
diamonds=pd.read_csv('diamonds.csv')
```

```python
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
sns.displot(diamonds['price'])
plt.show()
```

```python
sns.histplot(diamonds['price'])
plt.show()
```

```python
#when we don't want to see histogram then we write (hist=False)
```

```python
sns.distplot(diamonds['price'],hist=False)
plt.show()
```

```python
#change color
```

```python
sns.distplot(diamonds['price'],color='g')
plt.show()
```

```python
sns.distplot(diamonds['price'],color='pink')
plt.show()
```

```python
sns.distplot(diamonds['price'],color='purple')
plt.show()
```

```python
#want to only see histogram
```

```python
sns.distplot(diamonds['price'],color='red',bins=20,kde=False)
plt.show()
```

```python
sns.distplot(diamonds['price'],color='red',bins=10,kde=False)
plt.show()
```

```python
sns.distplot(diamonds['price'],color='red',bins=5,kde=False) #bins mean no of bars
plt.show()
```

```python
#vertical show
```

```python
sns.distplot(diamonds['price'],color='purple',vertical=True,bins=10,kde=False)
plt.show()
```

**BOX PLOT**

```python
churn=pd.read_csv('churn.csv')
```

```python
churn.head()
```

```python
sns.boxplot(x='Churn',y='tenure',data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn,color='g')
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,palette='Set1')
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,color='purple')
plt.show()
```

```python
#CHANGE LINEWIDTH THIKNESS
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,color='pink',linewidth=3)
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,color='pink',linewidth=3,order=['Two year','One year Contract','Month-to-month'])
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,hue='PaymentMethod')
plt.show()
```
